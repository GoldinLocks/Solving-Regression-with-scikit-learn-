```python
import pandas as pd
import sklearn
import autosklearn.classification as clsf
import autosklearn.regression as reg
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, explained_variance_score
```

```python
house_sales = pd.read_csv('kc_house_data.csv')
house_sales.info()
```

```python
house_sales['date'] = pd.to_datetime(house_sales.date).dt.strftime('%Y%m')
house_sales['date'] = house_sales['date'].astype(int)
house_sales.head()
```

```python
X = house_sales.drop('price', axis=1)
y = house_sales.price

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=804)
```

```python
auto_reg = reg.AutoSklearnRegressor(time_left_for_this_task=720, per_run_time_limit=40, 
                                    resampling_strategy='cv', resampling_strategy_arguments={'folds': 5})

auto_reg.fit(X_train.copy(), y_train.copy(), dataset_name='house_sales')

auto_reg.refit(X_train.copy(), y_train.copy())
```

```python
predictions = auto_reg.predict(X_test)
```

```python
print("Regressor r2 score =", explained_variance_score(y_true=y_test, y_pred=predictions))
print()
print(auto_reg.show_models())
print()
print(auto_reg.cv_results_())
print()
print(auto_reg.sprint_statistics())
```
